```python
# !pip install dabl
#Data cleaning
import dabl
import pandas as pd
df = pd.read_csv(r'C:\Users\MIT BIO\Desktop\Breast cancer.csv')
df_clean = dabl.clean(df)[::10]
df
#type of data
types = dabl.detect_types(df_clean)
print(types)
#Exploratory Data analysis¶
dabl.plot(df, target_col="info")
# Initial Model Building
ec = dabl.SimpleClassifier(random_state=0).fit(df, target_col="info")
#exaplin model
dabl.explain(ec)
```
